# Fine Tuning the ResNet50 Architecture for Image Classification

```python
# importing necessary packages
import torch
torch.manual_seed(10)
import torch.nn as nn
import torch.optim as optim
import numpy as np
import torchvision
from torchvision import datasets, models, transforms
import time
import os
from torchvision.models import resnet50, ResNet50_Weights
import matplotlib.pyplot as plt 
```

By treating the proven ResNet50 network as a baseline, a convolutional neural network for image classification can be constructed more efficiently and perform more accurately than a network built from scratch. Loading in the ResNet50 model and setting the weights to DEFAULT means that the model will be initialized with the best available pretrained weights for our version of torch. Since preprocessing our image data is necessary for achieving best performance, a preprocess step has been defined to include the preprocessing transforms associated with the ResNet50_Weights.DEFAULT weights. This transformation will be used to normalize image data going forward.

```python
# initializing model trained on ImageNet to fine tune
weights = ResNet50_Weights.DEFAULT
model = resnet50(weights = weights)

# defining corresponding image transforms
preprocess = weights.transforms()
```

The data used in this project was obtained via Kaggle (https://www.kaggle.com/datasets/gpiosenka/balls-image-classification) and contains images of 30 different varieties of balls including baseballs, crystal balls, wrecking balls, etc. The data is distributed across test, train, and valid set subfolders. Images are housed within folders of their corresponding types allowing for easy reference using Torchvision's ImageFolder class. Data within the train subfolder is read in and the preprocessing transforms are applied as mentioned earlier. The DataLoader is responsible for shuffling this data and separating it into smaller 10-image batches.

```python
# importing, applying image transforms, and preprocessing training data
train_data = datasets.ImageFolder("images/train", transform = preprocess)
trainloader = torch.utils.data.DataLoader(train_data, batch_size = 10, shuffle = True)
```

Each of the 30 types of balls can now be assigned for the model to reference. The classes attribute of train_data dataset allows for references to the class names which will be helpful when visualizing the model's output later on.

```python
# extracting class names
class_names = train_data.classes

# printing all class names
print(class_names)
```

While the image data is separated into 30 separate classes, ResNet50 does not necessarily have that same number of output features. In actuality, ResNet50 has 2048 output features which is significantly more than the 30 target features associated with the ball images dataset. In order to consolidate the model to output 30 features, a linear layer is applied as the final fully connected layer of our model. This layer takes the default number of output features from the ResNet50 model as input, applies a linear transformation, and returns a number of output features equivalent to the number of individual classes in the data.

```python
# ensuring required number of output features
feature_count = model.fc.in_features
model.fc = nn.Linear(feature_count, len(class_names))
```

In order to train the initialized model on the new training data, the function below is defined to assess and minimize the losses calculated between the target values fed to it and the predictions generated from it. The loss function and optimizer implemented are fairly standard and the training process does without a scheduled learning rate decay that other training functions may use. At the end of the training process, the trained model is saved locally for import or reference later.

```python
# defining loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr = 0.001, momentum = 0.9)

# defining model training process
def train_model(model, criterion, optimizer, num_epochs):
    
    training_start_time = time.time()
    
    model.train()
    
    for epoch in range(num_epochs):
        
        epoch_start_time = time.time()
    
        running_loss = 0.0
        for i, data in enumerate(trainloader, 0):
            inputs, labels = data

            optimizer.zero_grad()
        
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
        
            if i % 2000 == 1999:
                running_loss = 0.0
                
        print("- Epoch " + str(epoch) + " took {:.2f}s".format(time.time() - epoch_start_time))
            
    print("Training complete, process took {:.2f}s".format(time.time() - training_start_time))
    
    torch.save(model.state_dict(), "models/fine_tuning_model.pth")
    
    return
```

The model is fed to the training function, with the training data experiencing 5 forward passes and backpropogations through the network. These passes are each denoted by an epoch. Note that the training process for the model on this dataset took roughly two hours to complete.

```python
# training model
train_model(model, criterion, optimizer, num_epochs = 5)
```

The model has been fine-tuned on the training set and saved. In order to assess the performance of the model with the new trained parameters, the ResNet50 model is initialized from scratch and defined as before. Finally, the new weights that were saved are loaded and applied to the model and the test set images are loaded, preprocessed, and batched similarly to the training set. Unlike with the training set, it isn't necessary to shuffle the batches.

```python
# loading fine tuned model
weights = ResNet50_Weights.DEFAULT
model = resnet50(weights = weights)
feature_count = model.fc.in_features
model.fc = nn.Linear(feature_count, len(class_names))
preprocess = weights.transforms()
model.load_state_dict(torch.load("models/fine_tuning_model.pth"))
model.eval()

# importing, applying image transforms, and preprocessing testing data
test_data = datasets.ImageFolder("images/test", transform = preprocess)
testloader = torch.utils.data.DataLoader(test_data, batch_size = 10, shuffle = False)
```

In order to evaluate the performance of the model against the test set, the proportion of images classified correctly out of the total number of images is converted to a percentage. In this case the model correctly classified 98% of the images in the testing data. This is an exceptionally high performance rate that is achievable with the help of the ResNet50 architecture.

```python
# predicting across test set and calculating accuracy
correct = 0
total = 0
    
with torch.no_grad():
    for data in testloader:
        images, labels = data
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
```

```python
print(f"Accuracy on the test set is {100 * correct // total} %")
```

To visualize the model in practice, images from the valid dataset can be predicted and shown alongside their true classes and predicted classes. Since the images were normalized prior to feeding them into the model, they might appear discolored or distorted to the human eye. In order to convert the images back to their original form, the inverse of the transforms that formed the image preprocessing step can be applied to the processed images. After reversing the normalization step and assigning the true and predicted values to corresponding lists, the image, true value, and predicted value are shown for the final batch in the valid set. The model was able to correctly classify every image in this batch. Evidently, this model is very good at distinguishing different types of balls from each other.

```python
# reading in and predicting on validation data
valid_data = datasets.ImageFolder("images/valid", transform = preprocess)
validloader = torch.utils.data.DataLoader(valid_data, batch_size = 10, shuffle = True)

with torch.no_grad():
    for data in validloader:
        images, labels = data
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)

# reversing the mean and std normalization from preprocessing to better visualize images
inverse_preprocess = transforms.Compose([transforms.Normalize(mean = [0.0, 0.0, 0.0], std = [1/0.229, 1/0.224, 1/0.225]), transforms.Normalize(mean = [-0.485, -0.456, -0.406], std = [1.0, 1.0, 1.0])])
images = inverse_preprocess(images)

# displaying images and predictions for final batch
true_values = [class_names[label] for label in labels]
predicted_values = [class_names[prediction] for prediction in predicted]
for i in range(len(images)):
    plt.imshow(torchvision.utils.make_grid(np.transpose(images[i], (1, 2, 0))))
    plt.show()
    print("True: " + true_values[i] + ", Predicted: " + predicted_values[i])
```
